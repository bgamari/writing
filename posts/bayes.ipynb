```python
from __future__ import division
import numpy as np
from numpy import sqrt
import matplotlib.pyplot as pl
import scipy.stats
%matplotlib inline
```

```python
def fig(name):
    pl.savefig('../media/bayes/%s.svg' % name)
```

```python
np.random.seed(8)
```

```python
# Bernoulli parameter
alpha =  0.7

counts = [(0,0)]
obs = []
for i in range(100):
    n = np.random.random_sample()
    heads, tails = counts[-1]
    obs.append(int(n < alpha))
    if n < alpha:
        heads += 1
    else:
        tails += 1
    counts.append((heads,tails))
    
counts = np.array(counts)
```

```python
pl.plot(counts[:,0], counts[:,1])
pl.xlabel('number of tails')
pl.ylabel('number of heads')
fig('flips')
```

```python
def agresti_coull(n1, n, alpha):
    """ alpha here is the confidence interval """
    z = scipy.stats.norm.ppf(alpha/2)
    nn = n + z**2
    p = (n1 + z**2 / 2) / nn
    ci = z * sqrt(p * (1 - p) / nn)
    return (p, ci)
```

```python
ml = [agresti_coull(heads, heads+tails, 0.05)
      for (heads,tails) in counts[1:]]
ml = np.array(ml)
pl.errorbar(x=range(ml.shape[0]), y=ml[:,0], yerr=ml[:,1], fmt='.')
pl.axhline(alpha, c='0.2', lw=2)
pl.xlabel('Observation number')
pl.ylabel(r'Estimated $\alpha$')
fig('ml-inference')
```

```python
import scipy.special
def beta(a,b):
    return lambda x: x**(a-1) * (1-x)**(b-1) / scipy.special.beta(a,b)

def bayes(prior_counts):
    xs = np.linspace(0, 1, 300)
    for i in [0,1,2,4,10,20,40]:
        a,b = counts[i] + prior_counts
        pl.plot(xs, beta(a+1,b+1)(xs), label='%d flips' % i)

    handles, labels = pl.gca().get_legend_handles_labels()
    pl.gcf().legend(handles, labels, loc='right')
    pl.subplots_adjust(right=0.75)
    pl.xlabel(r'$\alpha$')
    pl.ylabel(r'Posterior probability')
```

```python
bayes([0,0])
fig('bayes')
```

```python
prior = [88,22]
prior = np.array(prior) / 4
bayes(prior)
fig('bayes-prior')
```
